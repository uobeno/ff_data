# What teams have had the highest performing season by total points? 

```python
%%sql
WITH prioritize_team AS (
    SELECT
        *,
        IIF(
            team LIKE '%TM',
            LAG(team) OVER(PARTITION BY c_player ORDER BY c_year ASC),
            COALESCE(team,adp_team)
        ) AS starting_team
    FROM nfl_adp_names
)
SELECT
    c_year,
    starting_team,
    ROUND(SUM(fantasy_points_ppr)) AS ppr_tot
FROM prioritize_team
WHERE 1=1
    AND fantasy_pos IN ('QB', 'WR', 'RB', 'TE')
    AND starting_team IS NOT NULL
GROUP BY 1, 2
ORDER BY 3 DESC
LIMIT 10
```

# How are those teams points split between position? 

```python
pos_split = """
WITH prioritize_team AS (
    SELECT
        *,
        IIF(
            team LIKE '%TM',
            LAG(team) OVER(PARTITION BY c_player ORDER BY c_year ASC),
            COALESCE(team,adp_team)
        ) AS starting_team
    FROM nfl_adp_names
),
team_tot AS (
    SELECT
        c_year,
        starting_team,
        ROUND(SUM(fantasy_points_ppr)) AS team_tot
    FROM prioritize_team
    WHERE 1=1
        AND fantasy_pos IN ('QB', 'WR', 'RB', 'TE')
        AND starting_team IS NOT NULL
    GROUP BY 1, 2
),
pos_tot AS (
    SELECT
        c_year,
        fantasy_pos,
        starting_team,
        ROUND(SUM(fantasy_points_ppr)) AS pos_tot
    FROM prioritize_team
    WHERE 1=1
        AND fantasy_pos IN ('QB', 'WR', 'RB', 'TE')
        AND starting_team IS NOT NULL
    GROUP BY 1, 2, 3
)
SELECT
    pos_tot.*,
    team_tot,
    ROUND(pos_tot / team_tot,2) AS pos_share
FROM pos_tot
LEFT JOIN team_tot
ON pos_tot.c_year = team_tot.c_year
AND pos_tot.starting_team = team_tot.starting_team
WHERE 1=1
    AND pos_tot.starting_team NOT LIKE '%TM'
    AND pos_tot IS NOT NULL
    AND pos_share IS NOT NULL
    AND team_tot > 500
"""
```

```python
%%sql
$pos_split
LIMIT 10
```

# Let's look at how that's split over time

```python
data = %sql $pos_split
df = pd.DataFrame(data)
```

```python
# Create the scatter plot
sns.set(style="whitegrid")

# Use FacetGrid to create a grid of plots based on fantasy_pos
g = sns.FacetGrid(df, row="fantasy_pos", hue="starting_team", height=4, aspect=2, palette='tab20')

# Map the scatterplot to the FacetGrid
g.map(sns.scatterplot, "team_tot", "pos_share", s=100, alpha=0.6)

# Add titles and labels
g.set_axis_labels("Total Team Points", "Position Share")
g.set_titles(row_template="{row_name}")

# Add legend
g.add_legend(title="Starting Team")

# Find top 20 pos_share values
top_pos_share = df.nlargest(20, 'pos_share')

# Find top 10 team_tot values
top_team_tot = df.nlargest(10, 'team_tot')

# Combine the top values
top_values = pd.concat([top_pos_share, top_team_tot]).drop_duplicates()

# Add data labels with offset and smaller size for filtered points
for ax in g.axes.flat:
    for _, row in top_values.iterrows():
        ax.text(row['team_tot'], row['pos_share'] + 0.0001, row['starting_team'], 
                horizontalalignment='left', size='small', color='black', weight='semibold')

# Show plot
plt.show()
```
