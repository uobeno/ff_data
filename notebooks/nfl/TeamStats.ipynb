```python
import os
from pathlib import Path
from sqlalchemy import create_engine
```

```python
# Define your database file path (SQLite example)
rel_db_path = Path.cwd().parent.parent / 'raw_data' / 'nfl' / 'src_code' / 'nfl.db'
db_file_path = rel_db_path.resolve()
engine = create_engine(f'sqlite:///{db_file_path}')
```

```python
%load_ext sql
```

```python
%sql sqlite:///{db_file_path}
```

```python
%%sql
SELECT
    year,
    cleaned_player,
    fantasy_pos,
    team,
    fppg,
    --rush_att_pg,
    --targets_pg,
    ROUND(fppg / SUM(fppg) OVER(PARTITION BY team, year),2) AS share_team_fppg,
    ROUND(fppg / SUM(fppg) OVER(PARTITION BY team, year, fantasy_pos),2) AS share_team_pos_fppg,
    next_fppg_growth
FROM nfl_enhanced
WHERE 1=1
    AND (fantasy_pos = 'WR' OR fantasy_pos = 'RB')
ORDER BY 5 DESC
LIMIT 200
```
